<h2>Monte Carlo numerical integration of $\int_0^{\pi } \sin (x) \, dx$</h2>

**Example** 
simple numerical integration

```python
import numpy as np

N = 100000

x = np.linspace(0, np.pi, N)
y = np.sin(x)

est_int = (np.pi/N)*sum(y)
est_int
```

<h3>Monte Carlo integration</h3>

```python
xp = np.random.rand(N)*np.pi
yp = np.random.rand(N)

ysin = np.sin(xp)
s = (yp < ysin)         # or just one line success = sum(yp < ysin)
success = sum(s)  

est_integral = (success/N)*np.pi
est_integral
```

**Plot to practice with some matplotlib functions**

```python
import matplotlib.pyplot as plt

x_d = np.linspace(0, np.pi, 1000)
y_d = np.sin(x_d)

plt.plot(x_d, y_d, c='r')
plt.scatter(xp, yp, c=['b' if s_i else 'k' for s_i in s])
```
