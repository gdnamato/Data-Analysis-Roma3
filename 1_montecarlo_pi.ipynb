
<h2>Estimating the value of pi using Monte Carlo Method</h2>


```python
import numpy as np

N = 1000000
x_p = np.random.rand(N)      # random np arrays of N point 
y_p = np.random.rand(N)

r = x_p**2 + y_p**2        # element by element operation (array of N elements)
s = sum(r<1)               # sum of the True values (as int 1)

est_pi = s/N*4
est_pi
```


**Plot**
